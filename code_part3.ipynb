## How can each category of review scores and housing type features be indicative of overall review ratings and whether a host is a superhost?

```python
# Encode the input Variables
def Encode(data):
    for column in data.columns[data.columns.isin(['host_is_superhost', 'property_type', 'room_type'])]:
        data[column] = data[column].factorize()[0]
    return data
listing = Encode(listing.copy())
```

```python
# The pairwise correlations between features
listing.corr()
```

## How to use various categories of review scores to predict overall review ratings for hosts?

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and test sets: 30% test set, 70% train set
X = listing[['review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 
            'review_scores_location', 'review_scores_value']]
y = listing['review_scores_rating']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn import linear_model

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr = regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)

# Plot outputs
x, y = pd.Series(y_test, name='Actual Review Scores Rating'), pd.Series(y_pred, name='Predicted Review Scores Rating')
sns.regplot(x, y, marker='x')
```

```python
# The coefficients
print('Coefficients: ', regr.coef_)
# The mean squared error
print('Mean squared error: %.2f' % np.mean((y_pred - y_test) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(X_test, y_test))
```

## How to use overall review ratings and housing types to predict whether a host is a super host?

```python
# Split the dataset into training and test sets: 30% test set, 70% train set
X = listing[['review_scores_rating', 'property_type', 'room_type']]
y = listing['host_is_superhost']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn import tree

# Train a supervised learning model that can be used to generate predicions for data
# Create classifier using DecisionTreeClassifier
clf = tree.DecisionTreeClassifier(max_depth=3)
clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from IPython.display import Image
import graphviz
import pydotplus

# Generate the decision tree
dot_data = tree.export_graphviz(clf, out_file=None, feature_names=['review_scores_rating', 'property_type', 'room_type'], class_names=['Host is superhost', 'Host is not superhost'], 
            filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
from sklearn import metrics

# Display model accuracy
print("Model accuracy:", metrics.accuracy_score(y_test, y_pred))

# Display model precision
print("Model precision:", metrics.precision_score(y_test, y_pred))

# Display model recall
print("Model recall:", metrics.recall_score(y_test, y_pred))
```

```python
# Generate classification report
target_names = ['Host is superhost', 'Host is not superhost']
print("Model classification report:")
print(metrics.classification_report(y_test, y_pred, target_names=target_names))
```

```python
print("Model confusion matrix:")
pd.DataFrame(metrics.confusion_matrix(y_test, y_pred), 
             columns=['Predicted: Host is superhost', 'Predicted: Host is not superhost'], 
             index=['Actual: Host is superhost', 'Actual: Host is not superhost'])
```
